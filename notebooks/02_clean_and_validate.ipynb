```python
from pathlib import Path
import pandas as pd
import numpy as np
```

```python
PROJECT_ROOT = Path.cwd().parent
RAW_DIR = PROJECT_ROOT / "data" / "raw"
PROCESSED_DIR = PROJECT_ROOT / "data" / "processed"

PROCESSED_DIR.mkdir(parents=True, exist_ok=True)

DATA_FILE = "WA_Fn-UseC_-Telco-Customer-Churn.csv"
DATA_PATH = RAW_DIR / DATA_FILE

DATA_PATH
```

```python
df = pd.read_csv(DATA_PATH)

df.shape
```

```python
df["TotalCharges"].head(10)
```

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ", np.nan)
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"])
```

```python
df["TotalCharges"].isna().sum()
```

```python
df["TotalCharges"] = df["TotalCharges"].fillna(0)
```

```python
df["TotalCharges"].isna().sum()
```

```python
df.dtypes
```

```python
yes_no_cols = []

for col in df.columns:
    if set(df[col].dropna().unique()).issubset({"Yes", "No"}):
        yes_no_cols.append(col)

yes_no_cols
```

```python
for col in yes_no_cols:
    df[col] = df[col].map({"Yes": 1, "No": 0})
```

```python
df[yes_no_cols].head()
```

```python
df.replace(
    {
        "No internet service": "No",
        "No phone service": "No"
    },
    inplace=True
)
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df.shape
df.head()
```

```python
clean_path = PROCESSED_DIR / "cleaned_churn_data.csv"
df.to_csv(clean_path, index=False)

clean_path
```

Hidden missing values in TotalCharges were detected and converted to numeric format. Missing values corresponded to new customers with zero tenure and were handled accordingly. Categorical values were standardized and validated to ensure analysis consistency.
