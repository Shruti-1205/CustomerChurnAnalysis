```python
DATA_FILE = "WA_Fn-UseC_-Telco-Customer-Churn.csv"
DATA_PATH = RAW_DIR / DATA_FILE

DATA_PATH.exists(), str(DATA_PATH)
```

```python
df = pd.read_csv(DATA_PATH)
df.shape
```

```python
display(df.head(10))
display(df.tail(5))
display(df.sample(10, random_state=42))
```

```python
overview = pd.DataFrame(
    {
        "column": df.columns,
        "dtype": [str(t) for t in df.dtypes],
        "missing_count": df.isna().sum().values,
        "missing_pct": (df.isna().mean().values * 100.0).round(2),
        "n_unique": df.nunique(dropna=True).values,
    }
).sort_values(by=["missing_count", "n_unique"], ascending=[False, True])

overview.reset_index(drop=True, inplace=True)
overview
```

```python
df.info()

duplicate_rows = df.duplicated().sum()
duplicate_rows
```

```python
def count_blank_like(series: pd.Series) -> int:
    if series.dtype != "object":
        return 0
    s = series.astype(str)
    return int((s.str.strip() == "").sum())

blank_like_counts = {col: count_blank_like(df[col]) for col in df.columns}
blank_like = (
    pd.DataFrame({"column": list(blank_like_counts.keys()), "blank_like_count": list(blank_like_counts.values())})
    .query("blank_like_count > 0")
    .sort_values("blank_like_count", ascending=False)
    .reset_index(drop=True)
)

blank_like
```

```python
possible_id_cols = [c for c in df.columns if "id" in c.lower()]
possible_id_cols
```

```python
target_candidates = [c for c in df.columns if c.lower() in ["churn", "target", "label", "y"]]
target_candidates
```

```python
cat_cols = [c for c in df.columns if df[c].dtype == "object"]

for c in cat_cols:
    vc = df[c].value_counts(dropna=False).head(15)
    print("\n" + "=" * 80)
    print("COLUMN:", c)
    display(vc)
```

```python
numeric_like_candidates = []
for c in df.columns:
    if df[c].dtype == "object":
        sample = df[c].dropna().astype(str).head(200)
        if len(sample) == 0:
            continue
        s = sample.str.replace(",", "", regex=False).str.strip()
        numeric_ratio = pd.to_numeric(s, errors="coerce").notna().mean()
        if numeric_ratio > 0.9:
            numeric_like_candidates.append((c, float(numeric_ratio)))

numeric_like_candidates
```

```python
if "Churn" in df.columns:
    churn_counts = df["Churn"].value_counts(dropna=False)
    display(churn_counts)

    plt.figure()
    churn_counts.plot(kind="bar")
    plt.title("Churn distribution")
    plt.xlabel("Churn")
    plt.ylabel("Count")
    plt.xticks(rotation=0)
    plt.show()
else:
    print("No 'Churn' column found. Columns are:", list(df.columns))
```

The dataset exhibited class imbalance with approximately 27% churn rate, which reflects realistic business churn patterns and informed model evaluation choices beyond accuracy.

```python
overview_path = REPORTS_DIR / "01_data_overview_columns.csv"
overview.to_csv(overview_path, index=False)
str(overview_path)
```

```python
notes = []

if duplicate_rows > 0:
    notes.append(f"Found duplicated rows: {duplicate_rows}")

if len(blank_like) > 0:
    notes.append("Found blank-like values (empty or whitespace) in object columns.")

if len(numeric_like_candidates) > 0:
    notes.append("Found object columns that appear numeric-like and may need type conversion: " +
                 ", ".join([x[0] for x in numeric_like_candidates]))

notes_path = REPORTS_DIR / "01_data_overview_notes.txt"
with open(notes_path, "w", encoding="utf-8") as f:
    if len(notes) == 0:
        f.write("No major issues detected in overview.\n")
    else:
        for line in notes:
            f.write("- " + line + "\n")

str(notes_path), notes
```
