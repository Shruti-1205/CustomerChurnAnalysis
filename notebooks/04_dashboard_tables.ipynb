```python
from pathlib import Path
import pandas as pd
```

```python
PROJECT_ROOT = Path.cwd().parent
PROCESSED_DIR = PROJECT_ROOT / "data" / "processed"

data_path = PROCESSED_DIR / "analytics_ready_churn_data.csv"

df = pd.read_csv(data_path)

df.shape
```

```python
kpi_summary = pd.DataFrame({
    "total_customers": [len(df)],
    "churn_rate_pct": [round(df["Churn"].mean() * 100, 2)],
    "total_revenue": [round(df["TotalCharges"].sum(), 2)],
    "avg_monthly_charge": [round(df["MonthlyCharges"].mean(), 2)]
})

kpi_summary
```

```python
churn_contract = (
    df.groupby("Contract")
      .agg(customers=("Churn","count"),
           churn_rate=("Churn","mean"))
      .reset_index()
)

churn_contract["churn_rate"] = churn_contract["churn_rate"] * 100
churn_contract
```

```python
churn_tenure = (
    df.groupby("tenure_group")
      .agg(customers=("Churn","count"),
           churn_rate=("Churn","mean"))
      .reset_index()
)

churn_tenure["churn_rate"] *= 100
churn_tenure
```

```python
churn_charges = (
    df.groupby("monthly_charge_group")
      .agg(customers=("Churn","count"),
           churn_rate=("Churn","mean"))
      .reset_index()
)

churn_charges["churn_rate"] *= 100
churn_charges
```

```python
churn_value = (
    df.groupby("customer_value_segment")
      .agg(customers=("Churn","count"),
           churn_rate=("Churn","mean"))
      .reset_index()
)

churn_value["churn_rate"] *= 100
churn_value
```

```python
revenue_risk = df[df["Churn"] == 1]["TotalCharges"].sum()
revenue_total = df["TotalCharges"].sum()

revenue_summary = pd.DataFrame({
    "total_revenue": [revenue_total],
    "revenue_at_risk": [revenue_risk],
    "percent_revenue_at_risk": [round((revenue_risk/revenue_total)*100,2)]
})

revenue_summary
```

```python
dashboard_dir = PROCESSED_DIR / "dashboard_tables"
dashboard_dir.mkdir(exist_ok=True)

kpi_summary.to_csv(dashboard_dir / "kpi_summary.csv", index=False)
churn_contract.to_csv(dashboard_dir / "churn_by_contract.csv", index=False)
churn_tenure.to_csv(dashboard_dir / "churn_by_tenure.csv", index=False)
churn_charges.to_csv(dashboard_dir / "churn_by_charge.csv", index=False)
churn_value.to_csv(dashboard_dir / "churn_by_value.csv", index=False)
revenue_summary.to_csv(dashboard_dir / "revenue_summary.csv", index=False)

dashboard_dir
```
